```python

# Task 1: Create a numpy array of size 2 x 2
import numpy as np

array_2x2 = np.array([[1, 2], [3, 4]])
print("2x2 Array:")
print(array_2x2)
```

```python

# Task 2: Create a random numpy array of size 3 x 4
random_3x4 = np.random.rand(3, 4)
print("Random 3x4 Array:")
print(random_3x4)
```

```python

# Task 3: Create numpy arrays of ones and zeroes
ones_array = np.ones((2, 3))
zeros_array = np.zeros((2, 3))

print("Array of Ones:")
print(ones_array)
print("\nArray of Zeroes:")
print(zeros_array)
```

```python

# Task 4: Create a random numpy array of size 6x3 and transpose it
rand_6x3 = np.random.randint(1, 10, (6, 3))
print("Original 6x3 Array:")
print(rand_6x3)

transposed = rand_6x3.T
print("\nTransposed Array (3x6):")
print(transposed)
```
